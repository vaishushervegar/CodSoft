# Train a Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = new_card.drop('Class',axis =1)
y = new_card['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model =  LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
predict_lg = log_model.predict(X_test)
```

```python
predict_lg 
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(y_test,predict_lg))
```

```python
print(classification_report(y_test,predict_lg))
print('\n')
print(confusion_matrix(y_test,predict_lg))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
predict_rfc = rfc.predict(X_test)
```

```python
print("Accuracy Score:",accuracy_score(y_test,predict_rfc))
```

```python
print(classification_report(y_test,predict_rfc))
print('\n')
print(confusion_matrix(y_test,predict_rfc))
```
