```python
def remove_null_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 35
        elif Pclass == 2:
            return 30
        else:
            return 25
    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply( remove_null_age,axis =1)
```

```python
sns.heatmap(data = df.isnull(),cmap = 'inferno',yticklabels = False)
```

```python
df.drop('Cabin',axis =1,inplace = True)
```

```python
sns.heatmap(data = df.isnull(),cmap = 'plasma',yticklabels = False)
```

```python
sex = pd.get_dummies(df['Sex'],drop_first=True)
embark = pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace = True)
```

```python
df = pd.concat([df,sex,embark],axis=1)
```

```python
df.head(5)
```

```python
df.set_index('PassengerId',inplace = True)
```

# Train a Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Survived',axis = 1)
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm_model = LinearRegression()
```

```python
lm_model.fit(X_train,y_train)
```

```python
predict_lm = lm_model.predict(X_test)
```

```python
print('Score:',lm_model.score(X_test,y_test))
```

```python
print('Coefficients:\n',lm_model.coef_)
```

```python
print('Intercept:\n',lm_model.intercept_)
```

```python
sns.set_palette('tab20c_r')
plt.scatter(y_test,predict_lm)
```

```python
sns.set_palette('viridis_r')
print(plt.hist(y_test-predict_lm,bins = 20))
```

# Model Evaluation

```python
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_test,predict_lm)
MSE = metrics.mean_squared_error(y_test,predict_lm)
RMSE = metrics.mean_squared_error(y_test,predict_lm)
print(f"MAE:{MAE}")
print(f"MSE:{MSE}")
print(f"RMSE:{RMSE}")
```
